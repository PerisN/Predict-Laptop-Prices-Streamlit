### **Extracting features using Regex**

```python
print(df.iloc[0,0])
print(df.iloc[0,-1])

print(df.iloc[1,0])
print(df.iloc[1,-1])
```

```python
# Extract features

def extract_features(row):
    product = row['Product']
    feature = row['Feature']
    
    ram_size = re.findall(r'(\d+) ?(?:GB|TB)', feature)
    ram_size = int(ram_size[0]) if ram_size else None
    
    ram_type = re.findall(r'(?:LP)?DDR\d\S*|Unified\sMemory', feature)
    ram_type = ram_type[0] if ram_type else None
    
    processor = re.findall(r'(?:AMD|Intel|M\d+|Qualcomm Snapdragon)[\s\w]*\b', feature)
    processor = processor[0] if processor else None

    storage = re.findall(r'(\d+) ?(?:GB|TB) ?(SSD|HDD)', feature)
    storage = [(f"{s[0]} {s[1]}") for s in storage] if storage else None
      
    os = re.findall(r'(Windows|Mac OS|Linux|DOS|Chrome)[\s\w]', feature)
    os = os[0] if os else None
    
    brand = re.findall(r'^\w+', product)
    brand = brand[0] if brand else None
    
    return pd.Series([ram_size, ram_type, processor, storage, os, brand],
                     index=['Ram Size', 'Ram Type', 'Processor', 'Storage', 'OS', 'Brand'])
    
df[['RAM_Size(GB)', 'RAM_Type', 'Processor', 'Storage(GB)', 'OS', 'Brand']] = df.apply(extract_features, axis=1)

df.sample(3)
```

```python
# Clean columns

def clean_storage(storage_str):
    # remove brackets and leading/trailing whitespace
    storage_str = storage_str.strip('[]').strip()

    # extract SSD value from the storage string using regex
    ssd_str = re.search(r'(\d+)\s*SSD', storage_str)

    # convert the extracted value to GB
    ssd_gb = int(ssd_str.group(1)) if ssd_str else 0

    return ssd_gb

def clean_processor(processor):
    return re.sub(r'Processor.*', '', str(processor))

df['Processor'] = df['Processor'].apply(clean_processor)
df['Storage(GB)'] = df['Storage(GB)'].astype(str)
df['Storage(GB)'] = df['Storage(GB)'].apply(clean_storage)

df.head()
```
